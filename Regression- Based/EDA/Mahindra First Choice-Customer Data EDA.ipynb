```python
dfCustomerMaster['Customer Partner Type Name'] = dfCustomerMaster['Partner Type']
partnerTypes = {'1':'Retail','2':'Corporate','3':'Fleet','4':'Employee','9001':'Insurance Company','9002':'Surveyor','9003':'Contact Person','-1':'Others'}
dfCustomerMaster['Customer Partner Type Name'] =dfCustomerMaster['Customer Partner Type Name'].fillna('-1')

dfCustomerMaster["Customer Partner Type Name"] = dfCustomerMaster["Customer Partner Type Name"].astype(float).astype(int).astype(str)
dfCustomerMaster['Customer Partner Type Name']= dfCustomerMaster['Customer Partner Type Name'].replace(partnerTypes)
dfCustomerMaster['Customer Partner Type Name'] =dfCustomerMaster['Customer Partner Type Name'].fillna('Others')
```

```python
dfCustomerMaster['Customer Partner Type Name'].value_counts()
```

```python
dataOriginDescription = {'Z001':'Camp - Outdoor','Z002':'Camp - Workshop','Z003':'Emailers','Z004':'Fleet','Z005':'Reference - Customer','Z006':'Reference - Employee','Z007':'Reference - Used car dealer','Z008':'Just Dial/Other helpline','Z009':'Snapdeal/Other websites','Z010':'Company website','Z011':'Float activity','Z012':'Petrol pump activity','Z013':'Hoardings/Outdoor Advertisements','Z014':'Insurance Company','Z015':'Television AD','Z016':'Newspaper AD','Z017':'Newspaper leaflet','Z018':'Outdoor Sales Activity','Z019':'Spotted the outlet','Z020':'Mahindra Sister concern Employee','Z021':'Other outdoor activity','Z022':'Radio'}
dfCustomerMaster['Customer Data Origin Description']= dfCustomerMaster['Data Origin']
dfCustomerMaster['Customer Data Origin Description']= dfCustomerMaster['Customer Data Origin Description'].replace(dataOriginDescription)
dfCustomerMaster['Customer Data Origin Description'] =dfCustomerMaster['Customer Data Origin Description'].fillna('Others')
```

```python
dfCustomerMaster['Customer Data Origin Description'].value_counts()
```

```python
dataTitle = {'1':'Ms.','2':'Mr.','3':'Company','4':'Mr. and Mrs.','-1':'Others'}
dfCustomerMaster['Title'] =dfCustomerMaster['Title'].fillna('-1')
dfCustomerMaster["Title"] = dfCustomerMaster["Title"].astype(float).astype(int).astype(str)
dfCustomerMaster['Customer Title']= dfCustomerMaster['Title']
dfCustomerMaster['Customer Title']= dfCustomerMaster['Customer Title'].replace(dataTitle)
dfCustomerMaster['Customer Title'] =dfCustomerMaster['Customer Title'].fillna('Others')
dfCustomerMaster
```

```python
dataMaritalStatus = {'1':'Single','2':'Married','3':'Widowed','4':'Divorced','5':'Seperated','-1':'Others'}
dfCustomerMaster['Marital Status'] =dfCustomerMaster['Marital Status'].fillna('-1')
dfCustomerMaster["Marital Status"] = dfCustomerMaster["Marital Status"].astype(float).astype(int).astype(str)
dfCustomerMaster['Customer Marital Status']= dfCustomerMaster['Marital Status']
dfCustomerMaster['Customer Marital Status']= dfCustomerMaster['Customer Marital Status'].replace(dataMaritalStatus)
dfCustomerMaster['Customer Marital Status'] =dfCustomerMaster['Customer Marital Status'].fillna('Others')
dfCustomerMaster
```

```python
dataOccupation = {'1':'Director','2':'Sr. Manager','3':'Vice President','4':'Assistant Vice President','5':'Senior Vice President','6':'Col','7':'Lt.','8':'Lt. Gen','9':'Doctor','10':'Capt','11':'Executive Director','12':'Executive Assistant',
                  '13':'Secretary','14':'Regional Manager','15':'Manager','16':'Junior Manager','17':'DGM','18':'General Manager','19':'Sr. General Manager','20':'Proprietor','21':'Partner','22':'Vice Chairman','23':'CFO',
                  '24':'CTO','25':'Self','26':'Managing Director','27':'Executive','28':'CEO','29':'Customer','30':'VIP Customer','31':'Others','-1':'Others'}
dfCustomerMaster['Occupation'] =dfCustomerMaster['Occupation'].fillna('-1')
dfCustomerMaster["Occupation"] = dfCustomerMaster["Occupation"].astype(float).astype(int).astype(str)
dfCustomerMaster['Customer Occupation']= dfCustomerMaster['Occupation']
dfCustomerMaster['Customer Occupation']= dfCustomerMaster['Customer Occupation'].replace(dataOccupation)
dfCustomerMaster['Customer Occupation'] =dfCustomerMaster['Customer Occupation'].fillna('Others')
dfCustomerMaster
```
