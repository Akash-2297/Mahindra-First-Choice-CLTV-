```python
#cleaning for Plant data
def clean_plant():
    plant.drop(["vendor_number_plant","plant"], axis=1, inplace=True)
    null_col = ["name_2", "house_number_and_street", "po_box", "sales_organization"]
    for col in null_col:
        plant[col].fillna("unknown", inplace=True)

clean_plant()
    
```

```python
for col in cat_col:
    plant[col].value_counts()[:10][::-1].plot.barh()
    for num, counts in enumerate(plant[col].value_counts()[:10][::-1]):
        plt.text(counts, num, counts)
    plt.title(col)
    plt.show()
```
